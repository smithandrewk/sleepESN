```python
from scipy import signal
from utils import *
import matplotlib.pyplot as plt
eeg = read_edf_by_fileindex(0).get_data(picks='EEG')[0]
```

```python
fs = 500
decimation_factor = 30
fs_decimated = int(fs/decimation_factor)
samples = len(eeg)
samples_decimated = int(samples/decimation_factor)
```

```python
x = np.linspace(0, len(eeg)/500/3600, samples, endpoint=False)
y = eeg
```

```python
ydem = signal.decimate(y, decimation_factor)
xnew = np.linspace(0, len(eeg)/500/3600, samples_decimated, endpoint=False)
```

```python
i = 1000
j = 1010
plt.figure(figsize=(16,9),dpi=100)
plt.plot(x[i*500:j*500], y[i*500:j*500], '-', xnew[i*fs_decimated:j*fs_decimated], ydem[i*fs_decimated:j*fs_decimated], '-')
plt.xlabel('Time, Seconds')
plt.legend(['data', 'decimated'], loc='best')
```
